```python
def minprodbin (k):
    i=1
    m=0
    while ((2*i)<=k):
        i=2*i
        m=m+1
    if i<k:
        t=1
        l=i
        while (t==1):
            if (l+(i/2))==k:
                m=m+1
                t=0
            elif (l+(i/2))<k:
                i=i/2
                l=l+i
                m=m+1
            else:
                i=i/2
    return m
```

```python
def minprodtrois(k):
    if k==3:
        m=2
    else:
        i=3
        m=2
        while ((2*i)<=k):
            i=2*i
            m=m+1
        if i<k:
            t=1
            l=i
            while (t==1):
                if (l+i/2)==k:
                    m=m+1
                    t=0
                elif (l+i/2)<k:
                    i=i/2
                    l=l+i
                    m=m+1
                else:
                    i=i/2
    return m
```

```python
def minprod(k):
    if k % 3 == 0 :
        m=minprodtrois(k)
    else :
        m=minprodbin(k)
    return m
```
